```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_load = "Resources/schools_complete.csv"
student_load = "Resources/students_complete.csv"
```

# District Summary

```python
# Read Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_load)
student_df = pd.read_csv(student_load)
```

```python
school_df.head()
```

```python
student_df.head()
```

```python
# Combine the data into a single dataset.  
school_complete_df = pd.merge(school_df, student_df, how="left", on=["school_name", "school_name"])
school_complete_df.head()
```

```python
#Data Collection for School data and Student Data
school_count = school_df["school_name"].count()
school_students = student_df["student_name"].count()
school_budget = school_df["budget"].sum()
average_math_score = student_df["math_score"].mean()
average_reading_score = student_df["reading_score"].mean()
passing_math = (school_complete_df["math_score"] >=70).sum()/school_students * 100
passing_reading = (school_complete_df["reading_score"] >=70).sum()/school_students * 100
pass_df = school_complete_df.loc[(school_complete_df["math_score"]>=70) & (school_complete_df["reading_score"]>=70)]

pass_df.head()
```

```python
pass_students = pass_df.shape[0]
```

```python
overall_percent = pass_students/school_students * 100
```

```python
summary_df = pd.DataFrame({"Total Schools": [school_count],
                           "Total Students": [school_students],
                           "Total Budget": [school_budget],
                           "Average Math Score": [average_math_score],
                           "Average Reading Score": [average_reading_score],
                           "% Passing Math": [passing_math],
                           "% Passing Reading": [passing_reading],
                           "% Overall Passing": [overall_percent]})

summary_df.style.format({"Total Students":"{:,}",
                         "Total Budget":"${:,.2f}"})
```

# School Summary

```python
students_count = school_complete_df["school_name"].value_counts()
students_count
```

```python
school_index_df = school_complete_df.groupby(["school_name"])
school_index_df.count()
```

```python
school_type = school_index_df["type"].first()
school_type
```

```python
school_passing_math = school_complete_df[(school_complete_df["math_score"] >= 70)]
school_passing_math
```

```python
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / students_count*100
per_school_passing_math
```

```python
school_passing_read = school_complete_df[(school_complete_df["reading_score"] >= 70)]
school_passing_read
```

```python
per_school_passing_reading = school_passing_read.groupby(["school_name"]).count()["student_name"] / students_count*100
per_school_passing_reading
```

```python
pass_both = school_complete_df.loc[(school_complete_df["math_score"]>=70) & (school_complete_df["reading_score"]>=70)]
```

```python
school_passing_overall = pass_both.groupby(["school_name"]).count()["student_name"] / students_count*100
school_passing_overall
```

```python
avg_math_scores = school_index_df["math_score"].mean()
avg_math_scores
```

```python
avg_read_scores = school_index_df["reading_score"].mean()
avg_read_scores
```

```python
#Data Collection for School data and Student Data
# school_budget_by_school = school_index_df["budget"].sum()
# per_school_budget = school_index_df["budget"].sum()/student_count_school
student_count_school = school_index_df["Student ID"].count()
per_school_budget = school_index_df["budget"].mean()
per_student_budget = per_school_budget/student_count_school
```

```python
summary_by_school_df = pd.DataFrame({"School Type":school_type,
                                     "Total Students":students_count,
                                     "Total School Budget":per_school_budget,
                                     "Per Student Budget":per_student_budget,
                                     "Average Math Score":avg_math_scores,
                                     "Average Reading Score":avg_read_scores,
                                     "% Passing Math": per_school_passing_math,
                                     "% Passing Reading": per_school_passing_reading,
                                     "% Overall Passing": school_passing_overall})

summary_by_school_df.style.format({"Total Students":"{:,}",
                         "Total School Budget":"${:,.2f}",
                         "Per Student Budget":"${:,.2f}",
                         "Average Math Score":"{:,.6f}",
                         "Average Reading Score":"{:,.6f}"})
                      
```

# Top Performing Schools (By % Overall Passing)

# Scores by School Spending

```python
spend_ranges = [0,584,629,644,675]
spend_labels = ["<$584","$585-629","$630-644","$645-675"]
```

```python
school_summary_copy["Spending Ranges (Per Student)"] = pd.cut(school_summary_copy["Per Student Budget"],spend_ranges,labels=spend_labels)
```
